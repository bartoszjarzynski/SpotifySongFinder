```python
# Import libraries
import pandas as pd
import numpy as np
# Import the Spotify CSV
spotify = pd.read_csv("Spotify-2000.csv")
# Display CSV data
spotify
```

```python
# Display columns with their max values
columns = spotify.columns[6::]
for col in columns:
    name = "max" + str(col) + " = "
    temp = spotify[col].max()
    print(name + str(temp))
```

```python
# Print columns and their types
print(columns)
type(columns)
df = spotify.copy()
del df["Length (Duration)"]
columns=df.columns[6::]
print(columns)
```

```python
# Normalizing each result - values between 1 and 100
for col in columns:
    if(col == "Loudness (dB)"):
        temp = spotify[col].min()
        spotify[col] = spotify[col]/temp * 100
        continue
    temp = spotify[col].max()
    spotify[col] = spotify[col]/temp * 100

spotify
```

```python
#dobieranie odpowiedniej piosenki
#porownac kazda wartosc wpisana przez uzytkownika z piosenkami
userchoice = [40,70,20,74,80,11,44,90]
df = spotify[["Energy", "Danceability","Loudness (dB)","Liveness","Valence","Acousticness","Speechiness","Popularity"]]
distancelist=[]
for i in range(len(df)):
    acc = []
    for j in range(len(userchoice)):
        acc.append(abs(userchoice[j] - df.iat[i,j])**len(userchoice))
    #acc = lista odleglosci punktu
    #pierwiastek stopnia acc z listy odleglosci(minkowski) - odleglosc punktu od wybranego przez uzytkownika
    temp = sum(acc)**(1/len(userchoice))
    distancelist.append(temp)
spotify["distance"] = distancelist
spotify
```

```python
#najlepsze piosenki to te z najmniejsza odlegloscia, sa najblizej wybranej przez uzytkownika

listofbestsings=spotify.nsmallest(3,["distance"])

print(listofbestsings["Title"])
```
